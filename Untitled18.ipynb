```python
import pandas as pd
import numpy as np
from scipy import stats
import datetime

# Load dataset
df = pd.read_csv('synthetic_healthcare_data.csv')

# 1. Detect Missing Values
def check_missing_values(df):
    missing = df.isnull().sum()
    print("Missing Values:")
    print(missing[missing > 0])

# 2. Detect Outliers using Z-score & IQR
def detect_outliers(df, column):
    z_scores = np.abs(stats.zscore(df[column].dropna()))
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_bound = df[column].quantile(0.75) + (1.5 * iqr)
    lower_bound = df[column].quantile(0.25) - (1.5 * iqr)
    outliers = df[(df[column] > upper_bound) | (df[column] < lower_bound)]
    print(f"Outliers in {column}:")
    print(outliers)

# 3. Check Data Type Inconsistencies
def check_data_types(df):
    print("Data Type Inconsistencies:")
    for col in df.columns:
        print(f"{col}: {df[col].dtype}")

# 4. Detect Temporal Anomalies
def check_temporal_anomalies(df):
    df['Admission_Date'] = pd.to_datetime(df['Admission_Date'], errors='coerce')
    df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'], errors='coerce')
    anomalies = df[df['Discharge_Date'] < df['Admission_Date']]
    print("Temporal Anomalies:")
    print(anomalies)

# 5. Detect Duplicate Records
def check_duplicates(df):
    duplicates = df[df.duplicated(subset=['Patient_ID'], keep=False)]
    print("Duplicate Records:")
    print(duplicates)

# Run Data Quality Checks
check_missing_values(df)
detect_outliers(df, 'Billing_Amount')
check_data_types(df)
check_temporal_anomalies(df)
check_duplicates(df)
```

```python
import numpy as np

# Convert dates to datetime format
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors="coerce")
df["Discharge_Date"] = pd.to_datetime(df["Discharge_Date"], errors="coerce")

# Find invalid date records
invalid_dates = df[df["Admission_Date"] > df["Discharge_Date"]]

# Scatter plot for admission and discharge dates
plt.figure(figsize=(10, 5))
plt.scatter(invalid_dates["Admission_Date"], invalid_dates["Discharge_Date"], color="red")
plt.xlabel("Admission Date")
plt.ylabel("Discharge Date")
plt.title("Patients with Incorrect Admission & Discharge Dates")
plt.xticks(rotation=45)
plt.show()
```
